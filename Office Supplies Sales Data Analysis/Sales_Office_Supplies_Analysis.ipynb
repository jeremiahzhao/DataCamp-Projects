```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
from plotly.subplots import make_subplots
import plotly.graph_objects as go
from datetime import datetime
import datetime as dt
from itertools import combinations
from collections import Counter
```

```python
df = pd.read_csv('office_supplies.csv')
df['Sales'] = round(df['Sales'],2)
df.rename(columns = {'Sales': 'Price', 'Quantity': 'Units Sold', 'Profit': 'Sales'}, inplace = True)
df.head()
```

```python
df['Sales'] = round(df['Price'] * df['Units Sold'] * (1- df['Discount']),2)
df
```

```python
sales = df.groupby(['Sub-Category', 'Region'])['Sales'].sum()
sales.to_frame()
```

```python
sales_regions = sales.unstack(level = 1)
sales_regions = sales_regions.reset_index()
sales_regions
```

```python
## What regions produce the most sales with each sub-category?

fig = px.bar(sales_regions, x="Sub-Category", y=['Central', 'East', 'South', 'West'], title="Sub-Category Sales by Region",
              width=900, height=600)

fig.update_layout(
    yaxis_title="Sales in USD ($)",
    legend_title="Region"
    )
fig.show()
```

```python
products = df.groupby(['Sub-Category', 'Region'])['Product Name'].count()
products.to_frame()
```

```python
products_regions = products.unstack(level = 1)
products_regions = products_regions.reset_index()
products_regions
```

```python
## Distribution of Products being bought in each region?

fig = px.bar(products_regions, x="Sub-Category", y=['Central', 'East', 'South', 'West'], title="Sub-Category Products by Region",
              width=900, height=600)

fig.update_layout(
    yaxis_title="Number of Products",
    legend_title="Region"
    )
fig.show()
```

```python
colors = ['gold', 'mediumturquoise', 'darkorange', 'lightgreen']

fig = go.Figure(data=[go.Pie(labels= df['Sub-Category'],
                             values=df['Sales'])])
fig.update_traces(textfont_size=10,
                  marker=dict(colors=colors, line=dict(color='#000000', width=2)))
fig.update_layout(
    title="Sales Revenue from Sub-Categories",
    legend_title="Region")
fig.show()
```

```python
df[["Year", "Month", "Day"]] = df["Order Date"].str.split("-", expand = True)
df
```

```python
monthly_sales = df.groupby(['Month', 'Region'])['Sales'].sum()
monthly_sales.to_frame()

monthly_sales = monthly_sales.unstack(level = 1)
monthly_sales = monthly_sales.reset_index()
monthly_sales
```

```python
monthDict={'01':'Jan', '02':'Feb', '03':'Mar', '04':'Apr', '05':'May', '06':'Jun', '07':'Jul', '08':'Aug', '09':'Sep', '10':'Oct', '11':'Nov', '12':'Dec'}
monthly_sales['Month'].replace(monthDict, inplace = True)
```

```python
fig = px.line(monthly_sales, x="Month", y=['Central', 'East', 'South', 'West'], title="Sales Revenue by Month on Average",
              width=900, height=600)

fig.update_layout(
    yaxis_title="Sales Revenue in USD ($)",
    legend_title="Region"
    )
fig.show()
```

```python
df['Order Date'] = df['Year'] + '-' + df['Month']
df['Order Date'] = pd.to_datetime(df['Order Date']).dt.strftime('%Y-%m')
df
```

```python
yearly_sales = df.groupby(['Order Date', 'Region'])['Sales'].sum()
yearly_sales.to_frame()

yearly_sales = yearly_sales.unstack(level = 1)
yearly_sales = yearly_sales.reset_index()
yearly_sales
```

```python


fig = make_subplots(rows=2, cols=2)

fig.add_trace(go.Scatter(x=yearly_sales['Order Date'], y= yearly_sales['Central'], name = 'Central'),
              row=1, col=1)

fig.add_trace(go.Scatter(x=yearly_sales['Order Date'], y= yearly_sales['East'], name = 'East'),
              row=1, col=2)

fig.add_trace(go.Scatter(x=yearly_sales['Order Date'], y= yearly_sales['South'], name = 'South'),
              row=2, col=1)

fig.add_trace(go.Scatter(x= yearly_sales['Order Date'], y= yearly_sales['West'],  name = 'West'),
              row=2, col=2)

fig.update_xaxes(title_text="Central Region Sales", row=1, col=1)
fig.update_xaxes(title_text="Eastern Region Sales", row=1, col=2)
fig.update_xaxes(title_text="Southern Region Sales", row=2, col=1)
fig.update_xaxes(title_text="Western Region Sales", row=2, col=2)

fig.update_yaxes(title_text="Sales Revenue in USD ($)", range=[0, 210000], row=1, col=1)
fig.update_yaxes(range=[0, 210000], row=1, col=2)
fig.update_yaxes(title_text="Sales Revenue in USD ($)", range=[0, 210000], row=2, col=1)
fig.update_yaxes(range=[0, 210000], row=2, col=2)

fig.update_layout(title_text="Sales Revenues Over the Years by Region", height=700)

fig.show()
```

```python
df1 = df[df['Order ID'].duplicated(keep=False)]
df1['Product Bundle'] = df1.groupby('Order ID')['Product Name'].transform(lambda x: '##'.join(x))

df1 = df1[['Order ID', 'Product Bundle']].drop_duplicates()
df1
```

```python
count = Counter()

for row in df1['Product Bundle']:
  row_list = row.split('##')
  count.update(Counter(combinations(row_list,2)))
print(count)
```

```python
count.most_common(10)
```

```python
product_group = df.groupby('Product Name').sum()
top20product = product_group.sort_values('Units Sold', ascending = False).head(20)
top20product = top20product.reset_index()
top20product = top20product.sort_values('Units Sold', ascending = True)
top20product['Product Name'] = top20product['Product Name'].str.split().str[:3].str.join(sep=" ")
top20product
```

```python
fig = px.bar(top20product, x="Product Name", y='Units Sold', title="Top 20 Products in Units Sold",
              width=900, height=600)

fig.update_traces(marker_color='rgb(158,202,225)', marker_line_color='rgb(8,48,107)',
                  marker_line_width=1.5, opacity=0.6)
fig.show()
```

```python
top20sales = product_group.sort_values('Sales', ascending = False).head(20)
top20sales = top20sales.reset_index()
top20sales = top20sales.sort_values('Sales', ascending = True)
top20sales['Product Name'] = top20sales['Product Name'].str.split().str[:5].str.join(sep=" ")
top20sales
```

```python
fig = px.bar(top20sales, x="Product Name", y='Sales', title="Top 20 Products in Units Sold",
              width=900, height=600)
fig.update_xaxes(tickangle=30)
fig.update_traces(marker_color='rgb(158,202,225)', marker_line_color='rgb(8,48,107)',
                  marker_line_width=1.5, opacity=0.6)
fig.show()
```
